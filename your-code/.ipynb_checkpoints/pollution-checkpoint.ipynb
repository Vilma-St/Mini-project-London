```python
import pandas as pd
emissions_summary = pd.read_excel('../../Mini-project-London/data/data-pollution/LAEI2016_Emissions_Summary-NOxPMCO2_modified.xlsx', sheet_name='MASTER')
emissions_summary
```

```python
education = pd.read_excel('../../Mini-project-London/data/data-education-modified/education-averages.xlsx', skiprows = 1, nrows = 35)
education = education.drop(labels=0)
education = education.drop(labels=34)
# df = df.rename(columns={'oldName1': 'newName1', 'oldName2': 'newName2'})
education = education.rename(columns = {"Area":"BOROUGHS"})
education
```

```python
import matplotlib.pyplot as plt

emissions_summary.plot(x = "BOROUGHS", y = ["2010 - NOx", "2013 - NOx", "2016 - NOx"], kind = "bar", figsize=(12,5))
plt.show()
```

```python
emissions_summary.plot(x = "BOROUGHS", y = ["2010 - PM10", "2013 - PM10", "2016 - PM10"], kind = "bar", figsize=(12,5))
plt.show()
```

```python
emissions_summary.plot(x = "BOROUGHS", y = ["2010 - PM2.5", "2013 - PM2.5", "2016 - PM2.5"], kind = "bar", figsize=(12,5))
plt.show()
```

```python
emissions_summary.plot(x = "BOROUGHS", y = ["2010 - CO2", "2013 - CO2", "2016 - CO2"], kind = "bar", figsize=(12,5))
plt.show()
```

```python
new = pd.merge(emissions_summary, education, on="BOROUGHS", how="left")
new
```

```python
#pandas.DataFrame.corrwith() - run correlations between columns in different data frames
# Corr() seems to work fine with merged data frame

correlation = new.corr(method ='pearson')
correlation
```

```python
schools_exced = pd.read_excel('../../Mini-project-London/data/data-pollution/Schools_exceeding_LAEI_2016_2-modified.xlsx', sheet_name='EdubaseAllEstablishmentsLondon')
schools_exced = schools_exced.rename(columns = {"Local Authority name":"BOROUGHS"})
schools_exced = schools_exced.groupby(['BOROUGHS']).mean()
schools_exced = schools_exced.drop(['URN'], axis=1)
schools_exced.head(60)
```

```python
new2 = pd.merge(schools_exced, education, on="BOROUGHS", how="left")
new2
```

```python
correlation2 = new2.corr(method ='pearson')
correlation2
```
