```python
education = pd.read_excel('../../Mini-project-London/data/data-education-modified/education-averages-to-import.xlsx', nrows = 35)
# education = education.drop(labels=0)
# education = education.drop(labels=34)
education = education.rename(columns = {"Area":"BOROUGHS"})
education
```

```python
import matplotlib.pyplot as plt

education.plot(x = "BOROUGHS",y = ["KS1 - % exp", "2010 - CO2", "2013 - CO2", "2016 - CO2"], kind = "bar", figsize=(12,4))
plt.show()
```
